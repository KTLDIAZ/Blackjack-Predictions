## Robust Regression

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X,y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3,40,1)
line_y_ransac = ransac.predict(line_X.reshape(-1,1))
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(12,8));
plt.scatter(X[inlier_mask], y[inlier_mask],
    c='blue', marker='o', label='inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
    c='gray', marker='s', label='outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('avg number of rooms')
plt.ylabel('medan value')
plt.legend(loc='upper left')
```

```python
print(ransac.estimator_.coef_)
print(ransac.estimator_.intercept_)
```

```python
yPredict = ransac.predict(X)
```

```python
r2_score(y, yPredict)
```

```python
ransac.score(X,y)
```

```python
line_X = np.arange(X.min(), X.max())[:, np.newaxis]
line_y = model.predict(line_X)
line_y_ransac = ransac.predict(line_X)
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(12,8));
plt.scatter(X[inlier_mask], y[inlier_mask],
    c='blue', marker='o', label='inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
    c='gray', marker='s', label='outliers')
plt.plot(line_X, line_y, color='purple', label='Linear regressor')
plt.plot(line_X, line_y_ransac, color='red', label="Ransac Regressor")
plt.xlabel('avg lstat')
plt.ylabel('medan value')
plt.legend(loc='upper left')
```

## Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
bdf.info()
```

```python
print(set(bdf['rad']))
print(set(bdf['chas']))
```

```python
feature_colums = ['crim', 'zn', 'indus', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat', 'medv']
x1 = bdf[feature_colums].values
y1 = bdf['chas'].values
```

```python
split_test_size = 0.30

x_train, x_test, y_train, y_test = train_test_split(x1,y1, test_size=split_test_size, random_state=1)
```

```python
num_obs = len(bdf)
num_true = len(bdf.loc[bdf['rad']==1])
print("BDF Chas True cases: {0} ({1:2.1f}%)".format(num_true,(num_true/num_obs)*100))
num_false = len(bdf.loc[bdf['rad']==0])
print("BDF chas False cases: {0} ({1:2.1f}%)".format(num_false,(num_false/num_obs)*100))

print("")
print("Training set True: {0} ({1:2.1f}%)".format(len(y_train[y_train[:]==1]), len(y_train[y_train[:]==1])/len(y_train)*100))
print("Training set False: {0} ({1:2.1f}%)".format(len(y_train[y_train[:]==0]), len(y_train[y_train[:]==0])/len(y_train)*100))

print("")
print("Test set True: {0} ({1:2.1f}%)".format(len(y_test[y_test[:]==1]), len(y_test[y_test[:]==1])/len(y_test)*100))
print("Test set False: {0} ({1:2.1f}%)".format(len(y_test[y_test[:]==0]), len(y_test[y_test[:]==0])/len(y_test)*100))
```

```python
len(x_train) 
```

## Inputs with the mean value

```python
from sklearn.impute import SimpleImputer

fill_0 = SimpleImputer(missing_values=0, strategy="mean")
x_train = fill_0.fit_transform(x_train)
x_test = fill_0.fit_transform(x_test)
```

## Training algorithm

```python
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()
nb_model.fit(x_train,y_train.ravel())
```

## Perfomance on training

```python
nb_predict_train = nb_model.predict(x_train)

from sklearn import metrics

print("Accuracy: {0:.2f}%".format(metrics.accuracy_score(y_train,nb_predict_train)*100))
```

```python
print("Confusion matrix")
print("{0}".format(metrics.confusion_matrix(y_train,nb_predict_train)))
```

## Perfomance on Test

```python
nb_predict_test = nb_model.predict(x_test)

print("Predictions: ", nb_predict_test)
```

```python
print("Accuracy: {0:.2f}%".format(metrics.accuracy_score(y_test, nb_predict_test)*100))
```

```python
print("Confusion matrix")
print("{0}".format(metrics.confusion_matrix(y_test,nb_predict_test)))
```
